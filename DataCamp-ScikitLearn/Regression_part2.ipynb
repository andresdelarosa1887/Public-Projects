# Ridge Regression

In summary, Ridge regression Penalizes  coefficients to avoid large coefficients. 

Models are penalized for coefficients that have large coefficients. 
alpha is a  parameter that controls  model complexity. 

- Loss function= OLS loss function + alpha * summartory of a^3
- Alpha: parameter that we need to choose 
- Picking alpha here is similar to picking k in k-NN
- Hyperparameter tuning 
- Alpha controls model complexity 
- Alpha=0: we get back OLS (can lead to overfitting)
- Very  high alpha: Can lead to underfiting


high alpha- large coefficients are significantly penalized - you could be underfitting the data 

low alpha- large coefficients are not so penalized so you could be overfitting the data
