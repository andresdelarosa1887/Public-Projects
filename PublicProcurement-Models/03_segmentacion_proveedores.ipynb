## Análisis de Componentes Principales (PCA)

```
##Partiendo los datos
X= X.drop('CLASIFICACION_EMPRESARIAL_2', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("datos de entrenamiento", X_train.shape)
```

```
##Normalizando los datos
sc = StandardScaler()  
X_train = sc.fit_transform(X_train)  
X_test = sc.transform(X_test)  
```

#### Steps for Principal Component Analysis

Eightenvalues calculation- 

- projecting the data onto the principal components 
- measuring the distances to the origin 
- Squaring and adding them together
- We can convert the  into variation around the origin ( 0,0) by dividing by the sample size minus 1 (i.e. n -1)


Example, 
Variation for PC1= 15
Variation for PC2 = 3
Total variation 15+ 3 =18
PC1 accounts for 83% of the total variation around the PCs. 


General steps
- Center the data 
- Find the best fitting line that goes through the origin
- For each factor we'd keep on finding more and more principal components by adding perpendicular lines and rotating them 
- With the eigenvalues for each principal component we determine the explained variance of each principal  component 

```
pca = PCA()  
X_train = pca.fit_transform(X_train)  
X_test = pca.transform(X_test)  
```

```
#########################
# Draw a scree plot and a PCA plot

#The following code constructs the Scree plot
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = [ str(x) for x in range(1, len(per_var)+1)]
plt.figure(figsize=(9, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('% de Varianza Explicada')
plt.xlabel('Componente Principal')
plt.title('Scree Plot de los datos de los Proveedores')
plt.show()
```

Selección de 15 componentes principales que toman en cuenta un % de la variacion 

```
pca= PCA(n_components=15)
X_train = pca.fit_transform(X_train) 
```

## Clustering de Proveedores

```
distorsion = []

for k in np.arange(1, 15):
    km = KMeans(n_clusters=k, 
                n_init=20, 
                max_iter=5000, 
                n_jobs=-1, 
                random_state=0)
    distorsion.append(km.fit(X_train).inertia_)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

diferencias = np.concatenate((np.array([1]),-np.diff(distorsion)))

plt.figure(figsize=(10, 8))
plt.plot(np.arange(1,15),distorsion,'--o',linewidth=2);
plt.plot(np.arange(1,15),diferencias,'-o')
plt.xticks(np.arange(1,15));
```
